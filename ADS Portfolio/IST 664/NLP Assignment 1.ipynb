# Data Preprocessing

## Lowercase all texts

```python
summaries=[x.lower() for x in summaries]
reviewtext=[x.lower() for x in reviewtext]
```

## Tokenization

```python
summary_tokens =[nltk.word_tokenize(i) for i in summaries]
review_tokens=[nltk.word_tokenize(i) for i in reviewtext] 
```

```python
from nltk.stem import WordNetLemmatizer
from nltk.corpus import wordnet
```

```python
nltk.download('wordnet')
nltk.download('omw-1.4')
wnl = nltk.WordNetLemmatizer()
```

##Remove Stop Words


```python
from nltk.corpus import stopwords
nltk.download('stopwords')
stop_words = set(stopwords.words('english'))
stop_words.update(['.', ',', '"', "'", '?', '!', ':', ';', '(', ')', '[', ']', '{', '}','#','...','&','"''"'])
```

```python
filtered_summaries=[]
for tokens in summary_tokens:
    for word in tokens :
        if not word in stop_words:
            filtered_summaries.append(word)
```

```python
filtered_reviews=[]
for tokens in review_tokens:
    for word in tokens :
        if not word in stop_words:
            filtered_reviews.append(word)
```

```python
print(len(filtered_summaries))
print(len(filtered_reviews))
```

##Lemmatization

```python
summary_lemma = [wnl.lemmatize(token) for token in filtered_summaries]
review_lemma = [wnl.lemmatize(token) for token in filtered_reviews]
```

# Data Analysis

## Bigrams with Highest Mutual Information

```python
from nltk.collocations import BigramAssocMeasures # We need two modules
from nltk.collocations import BigramCollocationFinder
bigram_measures = nltk.collocations.BigramAssocMeasures()
```

```python
findersummary = BigramCollocationFinder.from_words(filtered_summaries)
findersummary.apply_freq_filter(5)
scoredsummary = findersummary.score_ngrams(bigram_measures.pmi)
finderreview = BigramCollocationFinder.from_words(filtered_reviews)
finderreview.apply_freq_filter(5)
scoredreview = finderreview.score_ngrams(bigram_measures.pmi)
```

```python
for bscore in scoredreview[:50]:
    print (bscore)
```
